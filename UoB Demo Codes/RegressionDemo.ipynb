```python
##Harish Tayyar Madabushi - UoB Artifical Intelligence 1
import sys
import numpy as np
import matplotlib
matplotlib.use('TkAgg')
import matplotlib.pyplot as plt
```

```python
data = [[5, 10], [10, 19], [15, 21]]
```

```python
xs = [i[0] for i in data]
ys = [i[1] for i in data]
```

```python
x = np.linspace(min(xs), max(xs))
```

```python
epochs = 100
w1 = 0
w2 = 0
w0 = 0
alpha = 0.000001
for i in range(epochs):
    
    plt.clf()
    plt.scatter(xs, ys)
    plt.plot(x, w2 * x * x + w1 * x + w0)
    plt.ion()
    plt.show()
    plt.pause(0.000001)
    
    cost = 0
    for index in range(len(xs)):
        y_this = (w2 * xs[index] * xs[index]) + (w1 * xs[index]) + w0
        cost += (ys[index] - y_this) ** 2
        
        w2 += alpha * (ys[index] - y_this) * xs[index] * xs[index]
        w1 += alpha * (ys[index] -  y_this) * xs[index]
        w0 += alpha * (ys[index] - y_this)
        
    print("Cost, w0, w1, w2: ", (float(cost) / len(xs)), w0, w1, w2)
```
